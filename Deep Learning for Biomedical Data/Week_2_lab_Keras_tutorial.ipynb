## Warmup: Solving the XOR problem

First, we're going to go back and solve our XOR problem. Recall this is the data we were using:

```
import numpy as np

X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([0, 1, 1, 0]) 
```

Now we're going to create a very simple MLP to solve this problem.

```
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation
from tensorflow.keras.optimizers import SGD

# Your code here - build an MLP with one hidden layer
############################################################

num_features = X.shape[1]

model = Sequential()
model.add(Dense(units=32, activation='relu',
                input_dim=num_features, name='hidden_layer'))
model.add(Dense(units=1, activation='sigmoid', use_bias=False, name='output_layer')) 

opt = SGD(learning_rate=0.1)

model.compile(optimizer=opt, loss='mse', metrics=['accuracy'])
model.summary()
```

Now we fit the model:

```
model.fit(x=X, y=y, batch_size=1, epochs=1000)
```

## Introduction to Deep Learning in Keras

Now we are going to move on to some larger, more complex datasets where we actually need to use advanced deep learning concepts like GPUs, batching, etc and use the correct loss function (i.e. binary crossentropy) since we don't have to compute the gradients by hand :) 

We're first going to make a synthetic dataset:

```
# generate simulated data as in Lecture 1 Lab
from sklearn.datasets import make_classification

n_train = 2000
n_val = 500
n_classes = 2
X, y = make_classification(n_samples=n_train+n_val, 
                           n_features=10,
                           n_classes=n_classes,
                           n_informative=2, 
                           n_redundant=0,
                           random_state=1, 
                           shuffle=True)

X_train = X[0:n_train,:]
y_train = y[0:n_train]
X_val = X[n_train:(n_train+n_val),:]
y_val = y[n_train:(n_train+n_val)]

print(X_train.shape)
print(y_train.shape)
print(X_val.shape)
print(y_val.shape)
```
